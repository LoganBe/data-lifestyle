# Student Life Style and Stress - An Analysis
## Data collected from: https://www.kaggle.com/datasets/steve1215rogg/student-lifestyle-dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import seaborn as sns
from scipy.stats import permutation_test
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import r2_score, classification_report, confusion_matrix,ConfusionMatrixDisplay, accuracy_score
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LogisticRegression
```

```python
# Load in the data and examine data structure
df = pd.read_csv('student_lifestyle_dataset2.csv')
df.columns = df.columns.str.replace('_Hours_Per_Day', '', case=False) # Rename To Remove Hours_per_day
df = df.rename(columns={'Physical_Activity': 'Physical'}) # Rename to shorted Physical Activity
print(df.head())
print(df.dtypes)
print(df.isna().sum())

plt.rcParams.update({'font.size': 12})  # Set all font sizes to 14
```

## Examine Raw Data - Correlations and Distributions

```python
#%% Plot raw data | Distributions and correlations
df_floats = df.select_dtypes('float')
num_float = df_floats.shape[1]
plt.figure(1,figsize = (10,8)); plt.clf()
fig, axes = plt.subplots(num_float,num_float, num=1); 
plt.subplots_adjust(wspace=0.5, hspace=0.5)

norm = plt.Normalize(vmin=-1, vmax=1)
cmap = cm.get_cmap('coolwarm')

for i in range(num_float):
    data1_ = df_floats.iloc[:,i]
    for j in range(num_float):
        data2_ = df_floats.iloc[:,j]
        ax = axes[i, j]
        if i == j: # Diagnoal
            sns.histplot(data1_, ax=ax, kde=True, bins=20)
            ax.set_ylabel('');  ax.set_xlabel('')
        elif i < j:
            sns.scatterplot(x=data1_, y=data2_, ax=ax, s=10)
            ax.set_xlabel(''); ax.set_ylabel('')
        else:
            corr = data1_.corr(data2_)
            ax.set_facecolor(cmap(norm(corr)))
            ax.text(0.5, 0.5, f'{corr:.2f}', fontsize=12, ha='center', va='center')
            ax.set_xticks([])
            ax.set_yticks([])

        if i < 5 and i!=j:
            ax.set_xticklabels([])
        if j > 0:
            ax.set_yticklabels([])
        if i == 5:
            ax.set_xlabel(df_floats.columns[j])
        if j == 0:
            ax.set_ylabel(df_floats.columns[i])
plt.tight_layout()
```

## Make Pie charts showing distribution of factors and stress

```python
# Lets threshold our activties and see spread
# Go through all the data float data and create pie to show fraction (thresholded at 1/3)
plt.figure(3,figsize=(15, 8)); plt.clf()
fig, axs = plt.subplots(2, 3, num = 3)  # 3 rows, 2 cols
axs = axs.flatten()
explode = [0.05, 0.05, 0.05]  # same length as val_study

for nn in range(num_float):
    data_ = df_floats.iloc[:,nn].values
    min_ = np.min(data_); max_ = np.max(data_)
    range_ = max_-min_
    thresh_ = range_/3

    if df_floats.columns[nn] != 'GPA':
        label_ = [f"Small:\n {min_}-{min_+round(thresh_)} Hours",
                  f"Mid:\n {min_+round(thresh_)}-{(min_+round(2*thresh_,1))} Hours",
                  f"Large: {(min_+round(2*thresh_,1))}-{(min_+round(3*thresh_,1))} Hours"]
    else:
        #import pdb; pdb.set_trace()
        label_ = [f"Poor:\n {min_}-{min_+round(thresh_,2)}",
          f"Ok:\n {min_+round(thresh_,2)}-{(min_+round(2*thresh_,2))}",
          f"Good:\n {(min_+round(2*thresh_,2))}-{(min_+round(3*thresh_,2))}"]

    label_study = [None]*3
    val_study = [None]*3
    for i in range(3):
        label_study[i] = label_[i]
        val_study[i] = sum(data_[(data_ > min_+i*thresh_) & (data_ <= min_+(i+1)*thresh_)])
    
    axs[nn].pie(val_study,labels = label_study,autopct='%1.1f%%', explode=explode);
    if df_floats.columns[nn] != 'GPA':
        axs[nn].set_title(f"Fraction of {df_floats.columns[nn]} by Hours")
    else:
        axs[nn].set_title(f"Fraction of {df_floats.columns[nn]} by Score")
    axs[nn].set_aspect('equal')
plt.tight_layout()

#%% Make pie now for Stress
data_ = df['Stress_Level'].value_counts().values
label_stress = ['High Stress','Mid Stress','Low Stress']
plt.figure(4); plt.clf();
plt.pie(data_,labels = label_stress,autopct='%1.1f%%',explode = explode);
plt.title('Fraction of Stress Level')
```

## Find which factors significantly change with stress

```python
#%% Find Sig Between Activities and stress
acts = ['Study','Extracurricular','Sleep','Social','Physical','GPA']

# First lets reorganize so we have Stress, Activity, Value
df_melted = pd.melt(df, id_vars = 'Stress_Level'\
    ,value_vars = acts\
    ,var_name = 'activity',value_name = 'value')

# Remap value of streess to cateogrical values
stress_order = ['Low','Moderate','High']
df_melted['Stress_Level'] = pd.Categorical(df_melted['Stress_Level'], categories=stress_order, ordered=True)

stress_palette = {
    'Low':'#92C5DE',#'#56B1F7',
    'Moderate':'#F4A582', #'#FDB863',
    'High': '#D6604D' #'#CA0020'
}

# Make the box plots
plt.figure(2,figsize=(10, 8)); plt.clf()
ax = sns.boxplot(data = df_melted,x = 'activity',y='value',hue = 'Stress_Level',palette=stress_palette)

# Now lets get Significance
def diff_mean(x, y):
    return x.mean() - y.mean()

def pval_to_star(p):
    if p < 0.001: return '***'
    elif p < 0.01: return '**'
    elif p < 0.05: return '*'
    else: return 'ns'

# Set X values for Stars
positions = {'Low': 0, 'Moderate': 1, 'High': 2}
width = 0.1  # width shift between grouped boxes
offsets = {'Low': -0.25, 'Moderate': 0, 'High': 0.25}

# Go through the acts and get ***s - We will be using Permutation test
for i, metric in enumerate(acts):
    sub = df_melted[df_melted['activity'] == metric] # Get the activities

    # Get stress group values
    low = sub[sub['Stress_Level'] == 'Low']['value']
    mod = sub[sub['Stress_Level'] == 'Moderate']['value']
    high = sub[sub['Stress_Level'] == 'High']['value']

    # Compare ALL pairs within group across stress
    pairs = [('Low', 'Moderate'), ('Low', 'High'), ('Moderate', 'High')]
    for j, (g1, g2) in enumerate(pairs):
        x1 = i + offsets[g1]
        x2 = i + offsets[g2]
        y = sub['value'].max() + 1.5 + j  # adjust height
        # 10000 permutations
        temp = permutation_test((sub[sub['Stress_Level'] == g1]['value'],
                                sub[sub['Stress_Level'] == g2]['value']),
                                diff_mean, n_resamples=10000, alternative='two-sided')
        p = temp.pvalue # Extract pvalues
        star = pval_to_star(p) # Get star

        # Draw line and text
        ax.plot([x1, x1, x2, x2], [y, y + 0.1, y + 0.1, y], lw=1.2, c='black')
        ax.text((x1 + x2) / 2, y + 0.15, star, ha='center', va='bottom', fontsize=10)
        
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(title="Stress Level",loc='upper left')
#legend = ax.legend(frameon=False)
plt.ylabel('Self Reported Score (Hours or Grade)');
plt.xlabel('Activities');
```

## Linear Regression on GPA - Factors that contribute to grade

```python
acts = ['Study','Extracurricular','Sleep','Social','Physical','GPA']
model = smf.ols("GPA ~ Study + Extracurricular + Sleep + Social + Physical + C(Stress_Level)",data = df).fit()
summary_df = model.summary2().tables[1]

coefs = model.params
errors = model.bse
pvals = model.pvalues

stars = [pval_to_star(p) for p in pvals]

# Plot
plt.figure(3,figsize=(8, 6))
x_pos = np.arange(len(coefs))

plt.errorbar(x_pos, coefs, yerr=1.96*errors, fmt='o', color='black', capsize=5)
plt.axhline(0, color='gray', linestyle='--')

# Add stars above points
for i, (x, y, s, e) in enumerate(zip(x_pos, coefs, stars,errors)):
    offset = 1.96*e+0.02#
    plt.text(x, y + offset, s, ha='center', va='bottom', fontsize=10)

custom_labels = ['Stress High','Stress Low','Stress Mid','Study','Extracurricular','Sleep','Social','Physical']
plt.xticks(x_pos, custom_labels, rotation=45)
plt.ylabel("Coefficient Estimate")
plt.title("Effect of Predictors on GPA with Significance")
plt.ylim([-0.15,0.7])
plt.tight_layout()
```

## Predict GPA given factions

```python
X = df.drop(['GPA','Student_ID'],axis=1)
X = pd.get_dummies(X, drop_first=True)  # One-hot encode 'Stress'
y = df['GPA']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=0)
rf = RandomForestRegressor(n_estimators=100, random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print("R² Score on Test Set:", r2_score(y_test, y_pred))

result = permutation_importance(rf, X_test, y_test, n_repeats=30, random_state=0)
importances = pd.Series(result.importances_mean, index=X.columns)
print(importances.sort_values(ascending=False))

X = X.rename(columns={'Stress_Level_Low': 'Stress Low'})
X = X.rename(columns={'Stress_Level_Moderate': 'Stress Mid'})
X = X.rename(columns={'Stress_Level_Num': 'Stress High'})

# Create a DataFrame for nicer display
feat_importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': importances
}).sort_values(by='importance', ascending=False)

# Plot
plt.figure(4,figsize=(10,8)); plt.clf()
plt.bar(feat_importance_df['feature'], feat_importance_df['importance'])
plt.title('GPA Random Forest Feature Importances')
plt.ylabel('Importance')
ax = plt.gca()  # Get current axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#% Same but lets not include study as a variable
X = df.drop(['GPA','Student_ID','Study'],axis=1)
X = pd.get_dummies(X, drop_first=True)  # One-hot encode 'Stress'
y = df['GPA']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=0)
rf = RandomForestRegressor(n_estimators=100, random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print("R² Score on Test Set:", r2_score(y_test, y_pred))

result = permutation_importance(rf, X_test, y_test, n_repeats=30, random_state=0)
importances = pd.Series(result.importances_mean, index=X.columns)
print(importances.sort_values(ascending=False))

X = X.rename(columns={'Stress_Level_Low': 'Stress Low'})
X = X.rename(columns={'Stress_Level_Moderate': 'Stress Mid'})
X = X.rename(columns={'Stress_Level_Num': 'Stress High'})

# Create a DataFrame for nicer display
feat_importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': importances
}).sort_values(by='importance', ascending=False)

# Plot
plt.figure(5,figsize=(10,8)); plt.clf()
plt.bar(feat_importance_df['feature'], feat_importance_df['importance'])
plt.title('GPA Random Forest Feature Importances (Without Study)')
plt.ylabel('Importance')
plt.show()
ax = plt.gca()  # Get current axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

## Linear Regression on Stress

```python
#%% Linear Regression on Stress Level| No interactions - do not appear to help
acts = ['Study','Extracurricular','Sleep','Social','Physical','GPA']
stress_map = {'Low': 0, 'Moderate': 1, 'High': 2}
df['Stress_Level_Num'] = df['Stress_Level'].map(stress_map)
model = smf.ols("Stress_Level_Num ~ Study + Extracurricular + Sleep + Social + Physical",data = df).fit()
summary_df = model.summary2().tables[1]

coefs = model.params
errors = model.bse
pvals = model.pvalues

stars = [pval_to_star(p) for p in pvals]

# Plot
plt.figure(3,figsize=(8, 6))
x_pos = np.arange(len(coefs))

plt.errorbar(x_pos, coefs, yerr=1.96*errors, fmt='o', color='black', capsize=5)
plt.axhline(0, color='gray', linestyle='--')

# Add stars above points
for i, (x, y, s, e) in enumerate(zip(x_pos, coefs, stars,errors)):
    offset = 1.96*e+0.02#
    plt.text(x, y + offset, s, ha='center', va='bottom', fontsize=10)

custom_labels = ['Grade','Study','Extracurricular','Sleep','Social','Physical']
plt.xticks(x_pos, custom_labels, rotation=45)
plt.ylabel("Coefficient Estimate")
plt.title("Effect of Predictors on Stress with Significance")
plt.ylim([-0.2,0.7])
plt.tight_layout()


# Continuous predicted stress level
df['predicted_stress_cont'] = model.predict(df)
# Round and clip to valid range [0, 1, 2]
df['predicted_stress'] = df['predicted_stress_cont'].round().clip(0, 2).astype(int)

# Reverse the order of labels: High (2), Moderate (1), Low (0)
labels = [2, 1, 0]
label_names = ['High', 'Moderate', 'Low']

y_true = df['Stress_Level_Num']
y_pred = df['predicted_stress']

# Create confusion matrix with custom label order
cm = confusion_matrix(y_true, y_pred, labels=labels)

# Plot
plt.figure(12,figsize=(10, 8)); plt.clf()
ax = plt.gca()  # Get current axis
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_names)
disp.plot(cmap='Blues',ax=ax)
plt.title("Linear Regression Confusion Matrix")
plt.show()


# Accuracy
acc = accuracy_score(y_true, y_pred)
print(f"Rounded prediction accuracy: {acc:.2f}")
```

## Logistic Regression on Stress

```python
#%% Predict Stress given Variables - Lets do simple Logistic Regression First
# Prepare data
X = df.drop(['Stress_Level', 'Stress_Level_Num', 'Student_ID'], axis=1)
y = df['Stress_Level']  # categorical: Low, Moderate, High

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Initialize and fit logistic regression
log_clf = LogisticRegression(max_iter=1000, random_state=42, multi_class='multinomial', solver='lbfgs')
log_clf.fit(X_train, y_train)

# Predict
y_pred = log_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Prediction Accuracy: {accuracy:.2f}")

# Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5),num=43); plt.clf()
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=log_clf.classes_)
disp.plot(cmap='Blues', ax=plt.gca())
plt.title("Logistic Regression Confussion Matrix")

# Evaluate
print("Classification Report:\n", classification_report(y_test, y_pred))

# Now get importantce
importance = np.mean(np.abs(log_clf.coef_), axis=0)

# Create a sorted Series
importance_df = pd.Series(importance, index=X.columns).sort_values(ascending=False)
print(importance_df)

feat_importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': importance
}).sort_values(by='importance', ascending=False)

# Print sorted importances
print(feat_importance_df)

# Plot
plt.figure(313,figsize=(10,8)); plt.clf()
plt.bar(feat_importance_df['feature'], feat_importance_df['importance'])
plt.title('Logistic Regression Feature Importances')
plt.ylabel('Importance')
plt.show()
ax = plt.gca()  # Get current axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

## Random Tress on Stress

```python
#%% Predict Stress given Variables - Random Tree
# Features and target
X = df.drop(['Stress_Level','Stress_Level_Num','Student_ID'],axis=1)
y = df['Stress_Level']  # categorical: Low, Moderate, High

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Prediction Accuracy: {accuracy:.2f}")

# Confusion matrix
conf_mat = confusion_matrix(y_test,y_pred)
plt.figure(121,figsize=(6, 5)); plt.clf()
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=rf_clf.classes_)
disp.plot(cmap='Blues',ax=plt.gca())
plt.title("Random Forest Confusion Matrix")

# Evaluate
print("Classification Report:\n", classification_report(y_test, y_pred))

# Now get importantce
importances = rf_clf.feature_importances_

# Create a DataFrame for nicer display
feat_importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': importances
}).sort_values(by='importance', ascending=False)

# Print sorted importances
print(feat_importance_df)

# Plot
plt.figure(313,figsize=(10,8)); plt.clf()
plt.bar(feat_importance_df['feature'], feat_importance_df['importance'])
plt.title('Random Forest Feature Importances')
plt.ylabel('Importance')
plt.show()
```
